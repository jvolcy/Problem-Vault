```python
from bs4 import BeautifulSoup
import urllib
#from urllib.request import urlopen

def getPageSoupFromUrl(url):
    
    hdr = {'User-Agent':'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'} 
    #hdr = {'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8'}
    #Get the html data for the website
    
    request = urllib.request.Request(url,None,hdr) #The assembled request
    client = urllib.request.urlopen(request)
    htmlByteData = client.read()
    client.close()
    
    #print(htmlByteData)
    #Get and return the page soup
    pageSoup = BeautifulSoup(htmlByteData, "html.parser")
    return (pageSoup)
```

```python

def getVaultEntriesFromPageSoup(pageSoup):
    #create an empty problems list
    problems = []
    
    #pull every table row
    tr = pageSoup.findAll('tr')
    
    # for each row, look at the column data.  Valid rows will have the problem ID in column 0,
    # the problem description in column 1 and the problem URL in column 26
    for row in tr:
        td = row.findAll('td')
        #skip entries where the problem ID or the problem description is missing.
        if len(td[0].text)>1 and len(td[1].text)>1:
            #create a tuple and append to the problems list
            problems.append( (td[0].text, td[1].text, td[26].text) )
    
    return problems
    
```

```python
#URL of Google Problem Vault Sheet

# reading the spreadsheet URL directly through the standard shared link will only pull in 100 rows.  Also, a significant
# amount of scripting and other data accompanies the data.
# Per the website https://community.qlik.com/t5/QlikView-Scripting/Load-from-google-sheet-limited-to-100-rows/td-p/874206,
# it is possible to eliminate much of the packaged data and to get around the 100 row limit by using the template
# https://spreadsheets.google.com/tq?tqx=out:html&tq=&key=[DOC_ID]&gid=1
# where, [DOC_ID] (including braces) is replaced by the document ID in the standard URL.
# The usual vault shared URL is
# "https://docs.google.com/spreadsheets/d/1GoWDC0d23inlczi6EcaDGocmxm5-nBnXMVV_R5FDNlY/edit?usp=sharing"
# From this URL, the document ID is "1GoWDC0d23inlczi6EcaDGocmxm5-nBnXMVV_R5FDNlY"
# Using the template above, the URL we need is
# "https://spreadsheets.google.com/tq?tqx=out:html&tq=&key=1GoWDC0d23inlczi6EcaDGocmxm5-nBnXMVV_R5FDNlY&gid=1"
# 
#VAULT_URL = "https://docs.google.com/spreadsheets/d/1GoWDC0d23inlczi6EcaDGocmxm5-nBnXMVV_R5FDNlY/edit?usp=sharing"
VAULT_URL = "https://spreadsheets.google.com/tq?tqx=out:html&tq=&key=1GoWDC0d23inlczi6EcaDGocmxm5-nBnXMVV_R5FDNlY&gid=1"
```

```python
#pull the Problems Vault data from the vault URL (this may take some time)
vaultSoup = getPageSoupFromUrl(VAULT_URL)

# **** This seems to take several minutes to complete, but does eventually succeed. ****
```

```python
vault = getVaultEntriesFromPageSoup(vaultSoup)

for entry in vault:
    print(entry)
```
